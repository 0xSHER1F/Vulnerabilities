# **LFI (Local File Inclusion) :**

 * ```Occurs when a web application allows an attacker to include the contents of a local file on the server, it all about tricking The App into exposing or Running files on the Web Server```

     * It may lead to information disclosure, remote code execution (RCE)

 * **How it Works ?**

     * The App Uses Function like ( include - include_once - require - require_once ) to include a file based on a User-Controlled Variable , Example :

         ```php
         <?php
         $file = $_GET['file'];
         include('directory/' . $file);
         ?>
         ```

     * Attacker can manipulate this variable to point to a different file on the server :

         ```
         http://VulnerableSite.com/?file=../../../etc/passwd
         ```

# **LFD (Local File Disclosure) :**

 * ```specific type of LFI vulnerability where an attacker can access and view the contents of local files on the server``


# **LFI to LFD :**

 * Using PHP Stream Wrapper to Read the Contents of File 

 * **Example :**

     * ```X=php://filter/convert.base64-encode/resource=../../../..//etc/passwd```

         * This payload forces PHP to base64 encode the file before it is used or rendered in the response

         * The Previous PHP script attempts to include the file specified by the **X** value  

         * **php://filter :** 
           
             * meta-wrapper designed to permit the application of filters to a stream at the time of opening

         * **convert.base64-encode :** 
          
             * This is a filter that encodes data into base64 format

         * **resource :**

             * This specifies the resource to be filtered, which in this case is the file you want to read

     * In The Previous Example We Used Path Traversal to Access ``/etc/passwd`` Directory

# **LFI to RCE :**

 1) **PHP Session Files :**

     * Attackers can manipulate session files to inject malicious PHP code by controlling the session ID, they can include the poisoned session file, leading to RCE

     * The default location for these session files is **/var/lib/php/sessions**

 2) **Log Files With Controllable input :**

     * Some applications log sensitive information, including user input , If log files are writable, an attacker might be able to inject malicious code into them

     * **Examples :**

         * ``/var/log/apache2/access.log`` :

             * It's a detailed record of every request made to your website 

             * It Contains the Following Information :

                 1) **Remote host :** The IP address of the client making the request
                 2) **Remote user :** The username of the authenticated user 
                 3) **Date and time :** When the request was received
                 4) **Request method**
                 5) **Requested URL :** The resource requested by the client
                 6) **HTTP protocol version & Status Code**
                 7) **Bytes sent :** The number of bytes sent in the response
                 8) **Referrer :** The URL of the referring page
                 9) **User agent :** The browser or application used to make the request

         * ``/var/log/apache2/error.log`` :

             * It records detailed information about errors encountered during the operation of your website

             * It Contains the Following Information :

                 1) **Error messages :** Detailed explanations of errors that occurred request
                 2) **Script errors :** Issues with PHP, Perl, or other scripts 
                 3) **Configuration problems :** Errors related to Apache configuration files
                 4) **Access denied messages :** Attempts to access restricted files or directories
                 5) **System resource issues :** Problems with memory, CPU, or disk space
                 6) **Security incidents :** Possible attacks or suspicious activity

         * ``/var/log/vsftpd.log`` :

             * records a detailed history of FTP server activity, providing valuable insights into user actions, errors, and system behavior

             * It Contains :

                 * Successful and failed login attempts (Includes user names, IP addresses, and timestamps)

                 * Details about uploaded and downloaded files, including file names, sizes, and transfer times

                 * Data about incoming and outgoing connections, such as IP addresses and ports

                 * Records any issues encountered by the FTP server, such as authentication failures, file permission problems, or connection timeouts

         * ``/var/log/sshd.log`` :

             * records activities related to the SSH daemon (sshd) on a Linux system

             * provides detailed information about successful and failed login attempts, connection details, authentication methods, and other SSH-related events
     

 3) **Temporary Files :**

     * Uploaded files or temporary files can sometimes be exploited If these files are accessible through LFI, and the application processes them without proper validation, malicious code injection becomes possible

 4) **PHP Wrappers :**

     * PHP offers various wrappers **( php://input, php://filter )** that can be used to manipulate data streams , By combining these with LFI, attackers can potentially execute arbitrary code


 5) **Using file upload  forms\function :**

     * Attacker uploads a malicious script (e.g., a PHP shell) disguised as a legitimate file type (e.g., image, document)  

     * App has an LFI vulnerability that allows to include the contents of a local file , The attacker crafts a request to include the uploaded malicious script using the LFI vulnerability  

     * Web server executes the malicious code, granting the attacker remote control of the system

# **Path Traversal :** 

 * also known as **directory traversal**, is a vulnerability that allows an attacker to access files and directories outside of the web root folder 

 * Achieved by manipulating variables that reference files with "dot-dot-slash (../)" sequences and its variations

 * **How It Works ?**

     * Web App allows users to specify a file path as input , then uses this input to construct a file path to access the requested file

     * If the input is not properly validated or sanitized, an attacker can inject ``../`` sequences into the input to move up the directory structure , Which can lead to access of sensitive files like configuration files, source code, or system files


 * **Prevention :**

     * **Least privilege principle :** 
      
         * Run the application with minimal permissions

     * **Input validation :** 
     
         * Strictly validate and sanitize all user-supplied input      


     * **Whitelisting :** 
     
         * Only allow specific, predefined file paths  

     * **File path normalization :** 
     
         * Convert relative paths to absolute paths and remove any ".." or "." sequences 

     * **Restrict file system access :** 
     
         * Limit the application's ability to access certain directories
