# **File path traversal, validation of file extension with null byte bypass Lab :**

 * **Lab Objective :** Retrieve the contents of the ``/etc/passwd`` file 

 * **Given :** 
 
     * vulnerability in the display of product images
     * application validates that the supplied filename ends with the expected file extension

***

 * Open Burp's HTTP History , You Will Find Multiple Requests For Images Loading , Send any one of Them to The Repeater :

     * **From Givens :** application validates that the supplied filename ends with the expected file extension :

         * let's Use Null Byte Injection to Bypass the Validation :

             * **Null Byte injection :** 
             
                 * ```exploitation technique used to bypass sanity checking filters in web infrastructure by adding URL-encoded null byte characters (i.e. %00, or 0x00 in hex) to the user-supplied data```

                 * ```This injection process can manipulate the intended logic and behavior of the application and allow malicious attackers to get unauthorized access to the system files```


                 * In Our Case Null Byte Will Be Used As a Line Terminator , The App Will Check for the expected file extension (it Will Be There) But When Code is Executed the Null Byte Will Be the End Of The Line 

         * ```filename=../../../etc/passwd%00.jpg``` :

             * Directory Accessed successfully
