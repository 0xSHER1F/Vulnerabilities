# **Authentication :-**

 * **```The process of verifying the identity of a user, system, or entity before granting access to resources or performing actions, authentication ensures that only authorized users can access certain parts of the application or perform specific actions, such as viewing private data or making changes to a database```**


 * **Fundamental Types Of Authentication :**

     1) **Knowledge Factor (Something You Know) :**

         * This involves information that only the user should know 

         * **Examples :**

             * Passwords 
             * PINs
             * Security Questions 

     2) **Possession Factor (Something You Have) :**

         * This refers to a physical object in the user's possession that can be used for authentication


         * **Examples :**

             * Using SmartPhones for Receiving **two-factor authentication (2FA)** where a one-time passcode (OTP) is sent via SMS
             * Security Tokens 
             * Smart Cards
             * USB Security Keys         


     3) **Inherence Factor (Something You Are) :**

         * Involves using biometric data to verify the user's identity

         * **Example :**

             * Fingerprint Scanning 
             * Facial Recognition
             * Retina/Iris Scanning
             * Voice Recognition

***

 * **Difference Between Authentication & Authorization :**


     * **Authentication :** Verifies the identity of a user

     * **Authorization :** Determines what actions or resources an authenticated user is allowed to access.


     * **```Authentication = Who are you?```** &  **```Authorization = What can you do?```**

# **Multi-Factor Authentication (MFA) :**

 * **```Security mechanism that requires users to provide two or more verification factors to gain access to a system, application, or resource```**


 * **MFA Components :**

     * MFA involves a combination of the following authentication factors :

         1) **Knowledge Factor (Something You Know)**
         2) **Possession Factor (Something You Have)**
         3) **Inherence Factor (Something You Are)**


     * **Example :**

         * **SMS/Email-Based OTP :**

             * After entering a password, a one-time password (OTP) is sent to the user's phone via SMS/Email 

             * The user must enter this OTP to complete the login process


             * **```This Example Involves Combining Knowledge and Possession Factors , Something You Know (Password) & Something You Have (Smartphone)```**

# **Design Flaws in Authentication Mechanisms :-**

 * **Weak Password Policies :**

     * Allowing users to set simple, easily guessable passwords For Example : 

         1) Short or NULL Password
         2) Using Common Dictionary Words or Names
         3) Using Same Value of UserName as a Password

     * Weak passwords are vulnerable to brute-force attacks, credential stuffing, and dictionary attacks. If attackers can easily guess passwords, the security of the entire system is compromised  

     * **Mitigation :**

         * To prevent password-guessing attacks you may set a cookie such as failedlogins=1, and increment this following each unsuccessful attempt. When a certain threshold is reached, the server detects this in the submitted cookie and refuse to process the login attempt   


 * **Lack of Multi-Factor Authentication :**

     * Only requiring a single factor for authentication it can be easily bypassed if an attacker gains access to that factor, such as through phishing, keylogging, or password reuse attacks


 * **Verbose Failure Messages :**

     * Displaying Different Failure Message With  Revealing specific reasons for failure  can help attackers identify valid usernames 

     * **Example :**

         * If User Submitted an Invalid UserName a Message **```Incorrect Username```** is Displayed, When User Submits Valid UserName and Wrong Password a Message **```Incorrect Password```** 
          
             * Revealing Username Existence Could Lead to **Username Enumeration**  

             * Once an attacker identifies a valid username, they can combine it with common or leaked passwords to attempt brute-force attacks, trying to guess the correct password   

     * **Mitigation :**

         * Provide Generic Error Messages For Example **```"Authentication failed. Please check your username and password and try again."```**           



 * **Insecure Transmission of Credentials :**

     * Transmitting credentials over unencrypted channels (HTTP instead of HTTPS) can be intercepted by attackers through man-in-the-middle (MitM) attacks
