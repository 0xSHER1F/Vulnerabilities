# **SSRF (Server Side Request Forgery) :**

 * ```vulnerability that allows an attacker to manipulate a server into making requests to unintended locations (tricking a server into visiting a website of the attacker's choice)```

 * **Example :**

     * There is a Website that allows users to share links, Attacker can inject a URL pointing to an internal server instead of a public link . If the website fetches the content of the shared link, it might unintentionally reveal internal information or even allow the attacker to execute commands on the internal server 


 * **SSRF Types :**

     1) **Regular SSRF :**

         * Application returns the entire response of the forged request to the attacker

      2) **Blind SSRF :**

         * Application doesn't return the entire response of the forged request. Instead, it might indicate success or failure, or provide no feedback at all

 * **Protections against SSRF Attacks :**


     1) **Input Validation and Sanitization :**

     2) **Forbid unnecessary Internal Requests :**

     3) **Whitelisting and Blacklisting :**

         * **Blacklisting :**

             * Block access to known threat sources

         * **Whitelisting :**

             * Only permit requests to specific, trusted domains


         * **Whitelisting is generally more secure than blacklisting, as it's harder for attackers to bypass**

     4) **Network Restrictions :** 

         * **Utilize firewalls :** Implement network-level firewalls to control traffic

     5) **Defense in Depth :**

         * Employ multiple layers of protection

     6) **Regular review :** 
         
         * Continuously assess and update your security measures

 * **bypassing Techniques :**


     1) **Manipulate IP Address Formats :**

         * Using different IP address representations (decimal, hexadecimal, octal) can bypass filters 

     2) **URL Encoding :**

         * Encoding special characters can sometimes bypass input validation

     3) **Protocol Manipulation :**

         * Trying different protocols (ftp, file, gopher) might bypass restrictions   

     4) **Domain Pointing to localhost :**

 * **SSRF to LFD (Local File Inclusion) :**

     * Attacker can leverage SSRF to target local files on the server using LFI techniques  

     * **Exploitation Scenarios :**

         * **Reading sensitive files :**

             * By crafting a request to ``file:///etc/passwd`` or other sensitive files, an attacker can extract critical information


         * **Including malicious code :**

             * If the application dynamically includes the contents of the requested file, an attacker could upload a malicious script and then use SSRF to include it 


         * **Path traversal :**

             * By manipulating the file path, an attacker might be able to access files outside the intended directory

* **SSRF with Metadata :**

     * **Metadata :**

         * Metadata is data about data. In the context of cloud environments, metadata services provide information about the instance, such as its ID, region, availability zone, security groups, and potentially sensitive information like IAM credentials


     * By manipulating a vulnerable application to make a request to a metadata service, an attacker can potentially : 

         1) **Extract sensitive information :** Access IAM credentials, security group information, or other confidential data

         2) **Discover internal network topology :** Gather information about the instance's environment, potentially aiding in further attacks 

         3) **Lateral movement :** Use extracted credentials to access other resources within the cloud environment


     * **List of metadata service endpoints for different cloud providers :**

         ```https://github.com/Prinzhorn/cloud-metadata-services```
