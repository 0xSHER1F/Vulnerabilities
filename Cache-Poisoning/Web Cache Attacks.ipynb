# **Web Cache Poisoning :**

 * **Attack where an attacker manipulates a web cache to store harmful content, When other users subsequently request the same resource, they receive the poisoned content instead of the legitimate one**

 * **How it Works ?**

     1) **Identifying Unkeyed Headers :**

         * Attackers exploit unkeyed headers (headers not considered for cache key generation) to manipulate cache behavior

         * **Common Unkeyed Headers :**

             * **X-Forwarded-For**

             * **X-Forwarded-Host**

             * **Custom Headers :**

                 * Any custom headers added to requests might be unkeyed by default unless explicitly configured (depending on the caching server and its configuration) 

     2) **Crafting Malicious Requests :**

         * Crafting requests with specific unkeyed header values, the attacker can induce the origin server to generate a response containing malicious content 


     3) **Cache Manipulation :**

         * The attacker's malicious response is cached by the caching server due to identical cache keys


     4) **Serving Poisoned Content :**

         * Subsequent users requesting the same resource receive the poisoned content from the cache    


 * **What is The Impact of Web Cache Poisoning ?**

     * **XSS :** Injecting malicious JavaScript code into web pages

     * **Data Theft :** Stealing sensitive information   

     * **Redirect Attacks :** Redirecting users to phishing or malicious websites                

     * **Denial of Service (DoS) :** Overloading the cache with malicious content

# **Local Route Poisoning :**

 * **specific form of cache poisoning where an attacker manipulates the cache of a local network to redirect traffic to a malicious server**

 * **How it Works ?**

     1) **Identifying a Vulnerable Cache :** 
     
         * The attacker finds a cache server within the local network that is accessible and vulnerable to manipulation

     2) **Crafting a Poisoned Response :**

         * The attacker creates a malicious response that will be cached for a specific target URL

     3) **Injecting the Poisoned Response :**

         * The attacker injects the poisoned response into the cache, replacing the legitimate content

     4) **Redirecting Traffic :**

         * When users on the local network access the targeted URL, they receive the poisoned content, which can redirect them to a malicious site or serve malicious code

# **Web Cache Deception :**

 * **attack where an attacker tricks a web cache into storing sensitive, dynamic content that should not be cached. This is often achieved by manipulating the URL structure or exploiting misconfigurations in the caching server**

 * **How it Works ?**

     1) **Identifying Vulnerable URLs :** 
     
         * The attacker identifies dynamic URLs that contain sensitive information.

     2) **URL Manipulation :** 
     
         * By appending specific extensions or parameters to the URL, the attacker can trick the cache into thinking the content is static 


     3) **Cache Exploitation :** 
     
         * The cache stores the dynamic content as if it were a static resource       

     4) **Accessing Cached Data :** 
     
         * The attacker can then retrieve the cached content, potentially exposing sensitive information

 * **Example :**

     * Consider a dynamic web page, ```www.example.com/profile.php```, that displays a user's profile information . This page is not intended to be cached due to its dynamic nature

     * attacker might construct a URL like this ```www.example.com/profile.php/nonexistent.js``` :

         * Since ```.js``` typically represents a static file, the caching server might store the entire content of the profile page under the key associated with ```.js```

         * if the application is replaying with the sensitive user contents stored in ```www.example.com/profile.php```, you can steal those contents from other users
