# **Caching :**

 * **the process of storing copies of web resources (like HTML pages, images, and scripts) in a temporary storage area ```(cache)``` to improve website performance**

 * When a user requests a resource, the cache is checked first. If the resource is found in the cache, it's served from there, bypassing the origin server and significantly reducing load times

 * **How Web Caching Works ?**

     1) **Requesting Resources :** 

         * User requests a web page from a web server

     2) **Cache Checking :** 

         * Caching server checks its local cache for the requested resource   

     3) **Cache HIT :**

         * If the resource is found in the cache, it's served directly to the user

     4) **Cache MISS :** 

         * If the resource isn't in the cache, the request is forwarded to the origin server 

     5) **Caching :**

         * The fetched resource is stored in the cache for future requests, optimizing subsequent access

 * **Caching Servers :**

     * Store copies of frequently accessed web content,They act as intermediaries between clients (users) and origin servers (where the original content is stored)  

     * **Caching Servers Types :**

         1) **Web Proxies :** These act as intermediaries between clients and origin servers, caching web pages and other content, 
          
             * Examples : Squid, Nginx, and Varnish   

         2) **CDNs :** vast network of servers geographically distributed to deliver content quickly , CDNs excel at serving static content like images, videos, and CSS files 

         3) **Reverse Proxies :** While primarily used for load balancing, reverse proxies also cache content.

             * Examples : Nginx and HAProxy 

***

 * **Caching Key :**

     * Unique identifier used to store and retrieve cached data

     * It's essentially the fingerprint of a specific piece of content within a cache

     * When a user requests a resource, the caching server calculates a caching key based on various request parameters :

         * If a matching key exists in the cache, the cached content is served

         * otherwise, the content is fetched from the origin server and stored with the newly calculated key.


     * **Caching Key Components :**

         1) **URL :**

             * The complete URL of the requested resource, including the path and query parameters

         2) **HTTP Method :**

         3) **Headers :**

             * Certain HTTP headers, such as ```Accept```, ```Cookie```, or ```User-Agent```, can be included in the caching key to differentiate requests  

         4) **Query String :**

             * Parameters in the URL's query string can influence the caching key

         5) **Custom Parameters :** 
         
             * Some caching systems allow you to define custom parameters to include in the caching key               

 * **Unkeyed Headers :**

     * HTTP headers that are not included in the calculation of a caching key, their values don't affect whether a cached response is served or not

     * Unkeyed headers is a potential risk for exploitation in web **cache poisoning attacks** :

         * By manipulating unkeyed headers, an attacker can craft malicious requests that are cached and served to other users
