
# **XSS Cross-Site Scripting :-**

 * Web security vulnerability that allows attackers to inject malicious  scripts **(Payloads)** into trusted websites

     * > **For Payloads Visit =>** https://html5sec.org/ 

     * > **XSS Cheat Sheet =>** https://portswigger.net/web-security/cross-site-scripting/cheat-sheet    

 * >**Why Does it Happens ?**

     1) Trusting User's inputs, this User may be an Attacker and submits malicious code as input

     2) Bad Validations & Sanitizing to User's input , allowing the malicious code to be stored or reflected back to the user

***

 * >**Types Of XSS :**

     1) **Reflected XSS :**

         * Malicious code is sent to the website and immediately reflected back to the user, The application takes in user input, processes it server-side, and immediately returns it to the user

         * The attacker tricks the victim into clicking a link or visiting a page containing the reflected code 


     2) **Stored XSS :**

         * Malicious code is permanently stored on the website's server (in DB) and any user visiting the affected page will be vulnerable 


     3) **DOM-Based XSS :**

         * Malicious code manipulates the DOM of a web page after it's loaded  

         * similar to reflected XSS, except that in DOM-based XSS, the user input never leaves the user’s browser. In DOM-based XSS, the application takes in user input, processes it on the victim’s browser **(Client-Server)** and then returns it to the user  

     4) **Blind XSS :**

         * Malicious code is stored on the server, but the attacker cannot directly see its execution. The execution happens at a different time or location within the application, making it "blind" to the attacker  

         * **Example :**

             * **Admin Panel :**      

                 * attacker injects malicious code into a public form, but the script only executes when an administrator views that data in their dashboard


     5) **Self XSS :**

         * only affects the victim who injects the malicious code   

         * Attacker tricks the victim into submitting the malicious script, which then executes within the victim's own browser            

***

 * **Impacts Of XSS :**

     * Stealing session cookies to impersonate the victim and access their account 

     * Modifying the content of a website for malicious purposes

     * Stealing sensitive information like credit card details or login credentials

     * Redirecting users to fake login pages to steal their credentials

***

 * >**How To Mitigate XSS ?**

     1) Input Validation

     2) Output Escaping and Encoding **(Sanitizing)** 

         * **Escaping :** refers to the practice of encoding special characters so that they are interpreted literally instead of as a special character by the programs or machines that process the characters

     3) choosing the right JavaScript framework to use    

     4) Set the **HttpOnly** flag on sensitive cookies that your site uses, Which prevents attackers from stealing those cookies via XSS

     5) Implement the Content-Security-Policy HTTP response header, which restrict how resources such as JavaScript, CSS, or images load on your web pages

# **SOP (Same-Origin Policy) :-**

 * security mechanism in web browsers thatRestricts how documents or scripts loaded from one website (origin) can interact with resources from another website (origin)
 
 * it acts as a wall between websites, preventing malicious scripts from one site from accessing data on another site 

 * **When Two URLs are said to have the same origin ?**

     * if they share The Following :

         1) **Protocol**
         
         2) **Hostname**

         3) **Port Number**


     * **Example :** 

         * URL-A => https://example.com
         * URL-B => http://none.com
         * URL-X => https://example.com/@user


         * > A and X Share The Same Origin    

***

 # **CORS (Cross-Origin Resource Sharing) :**

 * Mechanism that builds upon the Same-Origin Policy (SOP) to allow controlled interactions between websites from different origins  

 * Enables websites to specify which origins (domains) are allowed to access their resources    

 * Uses HTTP headers to communicate permissions between website

***

# **CSP (Content Security Policy) :**

 * Works within a single origin to control what resources can be loaded by a web page       

 * Defines a set of trusted sources (domains, hashes) for scripts, images, etc

 * The browser only loads resources from the approved sources, mitigating risks from malicious code injection

 * https://csp-evaluator.withgoogle.com/

***
 

# **How SOP, CORS and CSP Work Together ?**


 * **SOP** establishes the base layer of security :
     
     * It restricts all cross-origin interactions by default

 * **CORS** creates exceptions :
     
     * Websites can define trusted origins for specific resources using CORS headers

 * **CSP** refines security within an origin : 
     
     * It dictates which resources a webpage can load from anywhere, including the website itself

# **HTTP Parameter Pollution (HPP) :**

 * Web security vulnerability that exploits how web applications handle user input within URLs or form submissions

 * It essentially involves manipulating or adding extra parameters to trick the application into unintended behavior

 * Attacker can send crafted HTTP requests with additional or modified parameters. This can potentially alter the application's logic or functionality

 * >**Common Techniques :**

     * **Repeating Parameters :**

         * Adding the same parameter name multiple times with different values, potentially affecting how the application handles the data

     * **Overriding Parameters :**

         * Including a parameter with the same name as a legitimate one, but with a malicious value to manipulate the application's logic.

     * **Spoofing Parameters :** 
     
         * Introducing entirely new parameters that the application might not expect, potentially leading to unexpected behavior
