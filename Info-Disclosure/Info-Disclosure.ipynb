# **Information Disclosure (Leakage) :**

 * ```a Vulnerability in Websites that allows unauthorized access to sensitive information or When The Website unintentionally reveals sensitive Information```

     * This information could be anything from personal data like usernames and passwords to confidential business documents or internal system details or Technical details about the website and its infrastructure

***

 * **Causes of Information Disclosure :**

     * **Error Messages (Tracing Enabled):** 
       
         * Verbose error messages might reveal details about the system, internal configurations, or even database structures 

     * Providing access to source code files via temporary backups

     * **Files for web crawlers :**

         * Revealing the names of hidden directories, their structure, and their contents via a ``( /robots.txt or /sitemap.xml )`` file or directory listing 


     * **Flawed Design and Coding :**

         * Programming errors or weaknesses in application logic can lead to unintended information leaks

     * **Developers Comments :**

         * Developers often leave comments within the code to explain functionality, document logic, or track progress these comments must be removed before deploying the code to a production environment   

         * developers forget or miss removing a comment during deployment, it can become accessible to anyone viewing the source code . Depending on the content of the comment, this can expose sensitive information             


***

 * **Impacts :**

     * **Data breaches :** 
     
         * Sensitive user or business data can be leaked, leading to identity theft, financial loss, or reputational damage  

     * **Increased attack surface :** 
     
         * Exposed information can be used by attackers to identify further vulnerabilities or plan more sophisticated attacks    


***

 * **Prevention :**

     * Implement strong access controls and authorization checks

     * Sanitize user input to prevent malicious code injection 

     * Use minimal logging and avoid storing sensitive information in logs

     * Regularly test for vulnerabilities and patch them promptly

     * Configure servers and applications securely
